    - –ü—Ä–∏–º–µ—Ä —Å —Å–∂–∞—Ç–∏–µ–º –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏—è

## –ó–∞–¥–∞—á–∞ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ (–∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏)

<img src="https://wiki.loginom.ru/images/linear-regression.svg" border="10px solid red"/>


>–ü—É—Å—Ç—å –∏–º–µ–µ–º –∫–æ–Ω–µ—á–Ω—ã–π –Ω–∞–±–æ—Ä —Ç–æ—á–µ–∫ $\left\{\left(x_i, y_i\right)\right\}_{i=1}^N$, –ø—Ä–∏—á—ë–º  $x_i \in \mathbb{R}^m, y_i \in \mathbb{R}$.
>
>__–•–æ—Ç–∏–º__ –Ω–∞–π—Ç–∏ __—Ñ—É–Ω–∫—Ü–∏—é__ $\check{y}$, __–Ω–∞–∏–ª—É—á—à–∏–º –æ–±—Ä–∞–∑–æ–º__ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É—é—â–∏–π –¥–∞–Ω–Ω—ã–π –Ω–∞–±–æ—Ä —Ç–æ—á–µ–∫.

–ü—Ä–æ—Å—Ç–µ–π—à–∏–π –ø—Ä–∏–º–µ—Ä - –ø—Ä–æ–≤–µ—Å—Ç–∏ –ø—Ä—è–º—É—é –ø–æ —Ç–æ—á–∫–∞–º (—Å–º. –∫–∞—Ä—Ç–∏–Ω–∫—É –≤—ã—à–µ).

–°—Ä–∞–∑—É –∂–µ –≤–æ–∑–Ω–∏–∫–∞–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø–æ–∏–Ω—Ç–æ–≤ –∫ –ø–æ—Å—Ç–∞–Ω–æ–≤–∫–µ –∑–∞–¥–∞—á–∏:

</br>

- __..., –Ω–∞–∏–ª—É—á—à–∏–º –æ–±—Ä–∞–∑–æ–º ...__ - –Ω—É–∂–µ–Ω —á–∏—Å–ª–æ–≤–æ–π –∫—Ä–∏—Ç–µ—Ä–∏–π "–∫–∞—á–µ—Å—Ç–≤–∞" –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏. –ò–º, –∫–∞–∫ –ø—Ä–∞–≤–∏–ª–æ, —Å–ª—É–∂–∏—Ç __—Ñ—É–Ω–∫—Ü–∏—è –æ—à–∏–±–∫–∏ (Loss-function).__

</br>

- __... —Ñ—É–Ω–∫—Ü–∏—é__ $\check{y}$, ...__ - –ø–µ—Ä–µ–±—Ä–∞—Ç—å –≤—Å–µ —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–µ–≤–æ–∑–º–æ–∂–Ω–æ (–º—ã —Ç—É—Ç –Ω–µ –Ω–∞ –º–∞—Ç–∞–Ω–µ), —Ç–∞–∫ —á—Ç–æ –ø—Ä–∏–¥—ë—Ç—Å—è —Å—É–∑–∏—Ç—å –∫—Ä—É–≥ –ø–æ–∏—Å–∫–∞ –Ω–∞ –Ω–µ–∫–æ—Ç–æ—Ä–æ–º __—Å–µ–º–µ–π—Å—Ç–≤–µ —Ñ—É–Ω–∫—Ü–∏–π.__

### –°–µ–º–µ–π—Å—Ç–≤–∞ —Ñ—É–Ω–∫—Ü–∏–π (–º–æ–¥–µ–ª–∏)

–ö–∞–∫ –º—ã —É–∂–µ –ø–æ–Ω—è–ª–∏, –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å –≤—Å–µ —Ñ—É–Ω–∫—Ü–∏–∏ –º—ã –Ω–µ –≤ —Å–∏–ª–∞—Ö –∏–∑-–∑–∞ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–π –∫–æ–º–ø—å—é—Ç–µ—Ä–∞. –ü–æ—ç—Ç–æ–º—É —Å—É–∑–∏–º –Ω–∞—à –∫—Ä—É–≥ –ø–æ–∏—Å–∫–∞ –Ω–∞ –∫–∞–∫–æ–µ-—Ç–æ —Å–µ–º–µ–π—Å—Ç–≤–æ —Ñ—É–Ω–∫—Ü–∏–π $\check{y}(x) = g_\mathbf{w}(x)$ (__–º–æ–¥–µ–ª–∏__), –∑–∞–≤–∏—Å—è—â–∏—Ö –æ—Ç –∫–∞–∫–æ–≥–æ-—Ç–æ –≤–µ–∫—Ç–æ—Ä–∞ —á–∏—Å–ª–æ–≤—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ (__–≤–µ—Å–æ–≤ –º–æ–¥–µ–ª–∏__) $\mathbf{w} = (\text{w}_1, \text{w}_2, ..., \text{w}_m)^\top$.

> –ù–∞—à–∞ —Ü–µ–ª—å –≤ –¥–∞–Ω–Ω–æ–º –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ - –Ω–∞–π—Ç–∏ __–æ—Å–æ–±–µ–Ω–Ω—ã–π__ –Ω–∞–±–æ—Ä –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ $\mathbf{w}^* = (\text{w}^*_1, \text{w}^*_2, ..., \text{w}^*_m)^\top$, –∫–æ—Ç–æ—Ä—ã–π –¥–∞—ë—Ç __–Ω–∞–∏–ª—É—á—à—É—é__ —Ñ—É–Ω–∫—Ü–∏—é $g_{\mathbf{w}^*}(x)$ –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ —Ç–æ—á–µ–∫ $\left\{\left(x_i, y_i\right)\right\}_{i=1}^N$.
>
> –≠—Ç–æ –∏ –±—É–¥–µ—Ç –Ω–∞—à–∞ –Ω–∞–∏–ª—É—á—à–∞—è –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏—è $\check{y}^*(x) \stackrel{\text{def}}{=} g_{\mathbf{w}^*}(x)$.

–ö–∞–∫ –≤—ã–±–∏—Ä–∞—Ç—å —Å–µ–º–µ–π—Å—Ç–≤–∞ —Ñ—É–Ω–∫—Ü–∏–π? –ï—Å—Ç—å –¥–≤–∞ –≥–ª–∞–≤–Ω—ã—Ö –ø–æ–¥—Ö–æ–¥–∞:

- –ò–∑ —Ç–µ–æ—Ä–∏–∏ (–Ω–∞—É–∫–∞) - –∑–Ω–∞–µ–º –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å, –≤–æ–ø—Ä–æ—Å –≤ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–∞—Ö (–∏—Ö –∏ "—Ñ–∏—Ç—Ç–∏–º").
- –ò–∑ –≥–æ–ª–æ–≤—ã (–∏—Å–∫—É—Å—Å—Ç–≤–æ) - –Ω–µ –∑–Ω–∞–µ–º –Ω–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏, –Ω–∏ –∫–æ–Ω—Å—Ç–∞–Ω—Ç. –°—Ç–∞—Ä–∞–µ–º—Å—è –ø–æ–¥–æ–±—Ä–∞—Ç—å –ø–æ–ª—É—á—à–µ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–∏.

- __–ü—Ä–∏–º–µ—Ä —Å –º–Ω–æ–≥–æ—á–ª–µ–Ω–∞–º–∏__

–ü—É—Å—Ç—å –µ—Å—Ç—å $\left\{\left(x_i, y_i\right)\right\}_{i=1}^N$ - –Ω–∞–±–æ—Ä —Ç–æ—á–µ–∫ –¥–ª—è "—Ñ–∏—Ç–∞". –ü–æ–∏—â–µ–º –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É—é—â—É—é —Ñ—É–Ω–∫—Ü–∏—é —Å—Ä–µ–¥–∏ __–º–Ω–æ–≥–æ—á–ª–µ–Ω–æ–≤ –≤—Ç–æ—Ä–æ–π —Å—Ç–µ–ø–µ–Ω–∏__:

$$
g_\mathbf{w}(x) = \text{w}_1 \cdot x^2 + \text{w}_2 \cdot x + \text{w}_3
$$

–ö–∞–∫ –≤–∏–¥–∏–º, –º–Ω–æ–∂–µ—Å—Ç–≤–æ —á–∏—Å–ª–æ–≤—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –ø—Ä–æ—Å—Ç–æ –≤–µ–∫—Ç–æ—Ä:

$$
\mathbf{w} = (\text{w}_1, \text{w}_2, \text{w}_3)^\top \in \mathbb{R}^3
$$

–≠—Ç–æ —á–∞—Å—Ç–Ω—ã–π —Å–ª—É—á–∞–π __–ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏__, —Ç.–∫. –ø—Ä–∏–∑–Ω–∞–∫–∏ (–≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –º–Ω–æ–≥–æ—á–ª–µ–Ω—ã) –≤—Ö–æ–¥—è—Ç –≤ –º–æ–¥–µ–ª—å –ª–∏–Ω–µ–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

plt.figure(figsize=(8, 6))

noise = 0.5*np.random.normal(size = 30)

def plot_data():

    x = np.linspace(-7, 7, 30)

    y = x**2 - 3*x - 5 + noise

    plt.scatter(x, y, c = 'r')


# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è –ø–∞—Ä–∞–±–æ–ª—ã
def plot_parabola(w1, w2, w3):

    plot_data()

    x = np.linspace(-10, 10, 400)
    y = w1 * x**2 + w2 * x + w3

    plt.plot(x, y, label=f'y = {w1}x^2 + {w2}x + {w3}', lw = 4)
    plt.title('$g_\mathbf{w}(x) = w_1 \cdot x^2 + w_2 \cdot x + w_3$')
    plt.xlabel('x')
    plt.ylabel('y')

    plt.xlim(-10, 10)
    plt.ylim(-10, 10)

    plt.grid(True)

    plt.show()



# –°–æ–∑–¥–∞–µ–º —Å–ª–∞–π–¥–µ—Ä—ã
w1_slider = widgets.FloatSlider(value=1.0, min=-10.0, max=10.0, step=0.1, description='w1')
w2_slider = widgets.FloatSlider(value=0.0, min=-10.0, max=10.0, step=0.1, description='w2')
w3_slider = widgets.FloatSlider(value=0.0, min=-10.0, max=10.0, step=0.1, description='w3')

# –°–æ–∑–¥–∞–µ–º –∏–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤
interact(plot_parabola, w1=w1_slider, w2=w2_slider, w3=w3_slider)
```

- __–ü—Ä–∏–º–µ—Ä —Å —Å—É–º–º–æ–π —Ä–∞–∑–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π__

–ü—É—Å—Ç—å –µ—Å—Ç—å $\left\{\left(x_i, y_i\right)\right\}_{i=1}^N$ - –Ω–∞–±–æ—Ä —Ç–æ—á–µ–∫ –¥–ª—è "—Ñ–∏—Ç–∞". –ü–æ–∏—â–µ–º –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É—é—â—É—é —Ñ—É–Ω–∫—Ü–∏—é —Å—Ä–µ–¥–∏ __—Å—É–º–º—ã –Ω–∞–æ–±—É–º –≤—ã–±—Ä–∞–Ω–Ω—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π__:

$$
g_\mathbf{w}(x) = \text{w}_1 \cdot f_1(x) + \text{w}_2 \cdot f_2(x)
$$

–≥–¥–µ

$$
f_1(x) = e^x, \quad f_2(x) = sin(x)
$$

–ö–∞–∫ –≤–∏–¥–∏–º, –º–Ω–æ–∂–µ—Å—Ç–≤–æ —á–∏—Å–ª–æ–≤—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –ø—Ä–æ—Å—Ç–æ –≤–µ–∫—Ç–æ—Ä:

$$
\mathbf{w} = (\text{w}_1, \text{w}_2)^\top \in \mathbb{R}^2
$$

–≠—Ç–æ —Ç–∞–∫–∂–µ —á–∞—Å—Ç–Ω—ã–π —Å–ª—É—á–∞–π __–ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏__, —Ç.–∫. –ø—Ä–∏–∑–Ω–∞–∫–∏ (–≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ —Ñ—É–Ω–∫—Ü–∏–∏ $f_1(x), f_2(x)$) –≤—Ö–æ–¥—è—Ç –≤ –º–æ–¥–µ–ª—å –ª–∏–Ω–µ–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

plt.figure(figsize=(8, 6))

noise = 0.5*np.random.normal(size = 30)

def plot_data():

    x = np.linspace(-7, 7, 30)

    y = 0.5*np.exp(x) + 3*np.sin(x) + noise

    plt.scatter(x, y, c = 'r')


# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è –ø–∞—Ä–∞–±–æ–ª—ã
def plot_model(w1, w2):

    plot_data()

    f1 = lambda x : np.exp(x)
    f2 = lambda x : np.sin(x)

    x = np.linspace(-10, 10, 400)
    y = w1 * f1(x) + w2*f2(x)

    plt.plot(x, y, lw = 4)
    plt.title('$g_\mathbf{w}(x) = w_1 \cdot f_1(x) + w_2 \cdot f_2(x)$')
    plt.xlabel('x')
    plt.ylabel('y')

    plt.xlim(-10, 10)
    plt.ylim(-10, 10)

    plt.grid(True)

    plt.show()



# –°–æ–∑–¥–∞–µ–º —Å–ª–∞–π–¥–µ—Ä—ã
w1_slider = widgets.FloatSlider(value=1.0, min=-10.0, max=10.0, step=0.1, description='w1')
w2_slider = widgets.FloatSlider(value=0.0, min=-10.0, max=10.0, step=0.1, description='w2')

# –°–æ–∑–¥–∞–µ–º –∏–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤
interact(plot_model, w1=w1_slider, w2=w2_slider)
```

- __–ü—Ä–∏–º–µ—Ä –Ω–µ–ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏__

–ü—É—Å—Ç—å –µ—Å—Ç—å $\left\{\left(x_i, y_i\right)\right\}_{i=1}^N$ - –Ω–∞–±–æ—Ä —Ç–æ—á–µ–∫ –¥–ª—è "—Ñ–∏—Ç–∞". –ü–æ–∏—â–µ–º –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É—é—â—É—é —Ñ—É–Ω–∫—Ü–∏—é —Å—Ä–µ–¥–∏ —Å–ª–µ–¥—É—é—â–µ–≥–æ —Å–µ–º–µ–π—Å—Ç–≤–∞ —Ñ—É–Ω–∫—Ü–∏–π:

$$
g_\mathbf{w}(x) = \frac{\sin(x^{\text{w}_1})}{2 + \cos(x^{\text{w}_2})}
$$

–ö–∞–∫ –≤–∏–¥–∏–º, –º–Ω–æ–∂–µ—Å—Ç–≤–æ —á–∏—Å–ª–æ–≤—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –ø—Ä–æ—Å—Ç–æ –≤–µ–∫—Ç–æ—Ä:

$$
\mathbf{w} = (\text{w}_1, \text{w}_2)^\top \in \mathbb{R}^2
$$

–í —Ç–æ –∂–µ –≤—Ä–µ–º—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è —Ç—É—Ç __–Ω–µ–ª–∏–Ω–µ–π–Ω–∞—è__ (–ø–æ—Ç–æ–º—É —á—Ç–æ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –º–æ–¥–µ–ª–∏ $\text{w}_i$ –Ω–µ –≤—Ö–æ–¥—è—Ç –≤ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ $g_\mathbf{w}(x)$ –ª–∏–Ω–µ–π–Ω—ã–º –æ–±—Ä–∞–∑–æ–º).

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

plt.figure(figsize=(8, 6))

noise = 0.05*np.random.normal(size = 30)

def plot_data():

    x = np.linspace(0, 5, 30)

    y = np.sin(x**2)/(2 + np.cos(x**1.5)) + noise

    plt.scatter(x, y, c = 'r')


# –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∏—Å–æ–≤–∞–Ω–∏—è –ø–∞—Ä–∞–±–æ–ª—ã
def plot_model(w1, w2):

    plot_data()

    x = np.linspace(0, 5, 2000)
    y = np.sin(x**w1)/(2 + np.cos(x**w2))

    plt.plot(x, y, lw = 2)
    plt.title('$g_\mathbf{w}(x) = (\sin(x^{w_1}))/(2 + \cos(x^{w_2}))$')
    plt.xlabel('x')
    plt.ylabel('y')

    plt.xlim(0, 5)
    plt.ylim(-2.5, 2.5)

    plt.grid(True)

    plt.show()



# –°–æ–∑–¥–∞–µ–º —Å–ª–∞–π–¥–µ—Ä—ã
w1_slider = widgets.FloatSlider(value=1.0, min=0.00001, max=10.0, step=0.1, description='w1')
w2_slider = widgets.FloatSlider(value=0.0, min=0.00001, max=10.0, step=0.1, description='w2')

# –°–æ–∑–¥–∞–µ–º –∏–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤
interact(plot_model, w1=w1_slider, w2=w2_slider)
```

## –§—É–Ω–∫—Ü–∏—è –æ—à–∏–±–∫–∏ (Loss-function)

–í—ã–±–∏—Ä–∞—Ç—å –Ω–∞–∏–ª—É—á—à—É—é –º–æ–¥–µ–ª—å "–Ω–∞ –≥–ª–∞–∑", –∫–æ–Ω–µ—á–Ω–æ, –≤–µ—Å–µ–ª–æ, –Ω–æ –Ω–µ –æ—á–µ–Ω—å –ø—Ä–∞–∫—Ç–∏—á–Ω–æ. –ù—É–∂–µ–Ω —á–∏—Å–ª–æ–≤–æ–π –∫—Ä–∏—Ç–µ—Ä–∏–π "–∫–∞—á–µ—Å—Ç–≤–∞" –º–æ–¥–µ–ª–∏ –¥–ª—è  –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É–µ–º–æ–≥–æ –Ω–∞–±–æ—Ä–∞ $\left\{\left(x_i, y_i\right)\right\}_{i=1}^N$.

–ù–∞—á–Ω—ë–º —Å —Å–∞–º–æ–≥–æ –ø—Ä–æ—Å—Ç–æ–≥–æ - __—Ñ—É–Ω–∫—Ü–∏–∏ –æ—à–∏–±–∫–∏ (Loss-function)__ $\mathcal{L}$. –ö–∞–∫ —Å–ª–µ–¥—É–µ—Ç –∏–∑ –Ω–∞–∑–≤–∞–Ω–∏—è - __—á–µ–º –±–æ–ª—å—à–µ —ç—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è, —Ç–µ–º —Ö—É–∂–µ –º–æ–¥–µ–ª—å__ (–∏ –Ω–∞–º –Ω–∞–¥–æ –µ—ë –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å, —É–¥–∏–≤–∏—Ç–µ–ª—å–Ω–æ, –¥–∞).

–ï—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω–æ, –Ω–∞ –≤—Ö–æ–¥ —ç—Ç–∞ —Ñ—É–Ω–∫—Ü–∏—è –¥–æ–ª–∂–Ω–∞ –ø—Ä–∏–Ω–∏–º–∞—Ç—å –∫–∞–∫ –∏—Å—Å–ª–µ–¥—É–µ–º—É—é –º–æ–¥–µ–ª—å $\check{y}(x) = g_\mathbf{w}(x)$, —Ç–∞–∫ –∏ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É–µ–º—ã–π –Ω–∞–±–æ—Ä $\left\{\left(x_i, y_i\right)\right\}_{i=1}^N$. –ü–æ–ª—É—á–∞–µ—Ç—Å—è –∫–∞–∫-—Ç–æ —Å–ª–æ–∂–Ω–æ, –ø–æ—ç—Ç–æ–º—É –≤–≤–µ–¥—ë–º –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω—É—é —Ç–µ—Ä–º–∏–Ω–æ–ª–æ–≥–∏—é, —É–ø—Ä–æ—â–∞—é—â—É—é —Å–∏—Ç—É–∞—Ü–∏—é.

>__–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ.__ –ù–∞–∑–æ–≤—ë–º $\check{y}_i = \check{y}(x_i) = g_\mathbf{w}(x_i)$ __–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è–º–∏__ –º–æ–¥–µ–ª–∏ –Ω–∞ –∏—Å—Å–ª–µ–¥—É–µ–º–æ–º –Ω–∞–±–æ—Ä–µ.

–ê —Ç–æ–≥–¥–∞ —Ñ—É–Ω–∫—Ü–∏—è –æ—à–∏–±–∫–∏ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ —á–∏—Å—Ç—ã–µ –¥–∞–Ω–Ω—ã–µ (__label__ - $y_i$) –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ (__predict__ - $\check{y}_i$). –ò –Ω–∞—à–∞ –≥–ª–æ–±–∞–ª—å–Ω–∞—è —Ü–µ–ª—å - –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –æ—à–∏–±–∫–∏ (–∞ —Ç–æ—á–Ω–µ–µ –Ω–∞–π—Ç–∏ $\mathbf{w}^*$, —Ç–æ –µ—Å—Ç—å –∞—Ä–≥—É–º–µ–Ω—Ç, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–º $\mathcal{L}$ –º–∏–Ω–∏–º–∏–∑–∏—Ä—É–µ—Ç—Å—è):

$$
\mathbf{w}^* = \text{arg}\min_\mathbf{w} \mathcal{L} (y_i, \check{y}_i)
$$


–ù–µ–∫–æ—Ç–æ—Ä—ã–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –æ—à–∏–±–∫–∏:
    
- __MSE - Mean Square Error (–ú–ù–ö)__

$$
\mathcal{L}\left(\check{y}_i, y_i\right)=\frac{1}{N} \sum_{i=1}^N\left(\check{y}_i-y_i\right)^2=\frac{1}{N} \sum_{i=1}^N\left(g_\mathbf{w}\left(x_i\right)-y_i\right)^2
$$

–Ø–≤–ª—è–µ—Ç—Å—è —Å–∞–º–æ–π —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–æ–Ω—ë–Ω–Ω–æ–π –∏ –ø—Ä–æ—Å—Ç–æ–π –º–µ—Ç—Ä–∏–∫–æ–π –æ—à–∏–±–æ–∫. –õ–µ–≥–∫–æ –ø–æ–ª—É—á–∞—Ç—å –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –∏ —á–∏—Å–ª–µ–Ω–Ω–æ –º–∏–Ω–∏–º–∏–∑–∏—Ä–æ–≤–∞—Ç—å.

- __RMSE - Root Mean Square Error__

$$
\mathcal{L}\left(\check{y}_i, y_i\right)=\sqrt{M S E}=\sqrt{\frac{1}{N} \sum_{i=1}^N\left(g_\mathbf{w}\left(x_i\right)-y_i\right)^2}
$$

- __Cross-Entropy, log-loss__

–ò—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –≤ –∑–∞–¥–∞—á–µ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ (–∫–æ–≥–¥–∞ $y_i \in \{0, 1\}$):

$$
\mathcal{L}\left(\check{y}_i, y_i\right)=-\frac{1}{N} \sum_{i=1}^N\left(y_i \cdot \log \left(\check{y}_i\right)+\left(1-y_i\right) \cdot \log \left(1-\check{y}_i\right)\right)
$$

- __...__

–í –¥–∞–Ω–Ω–æ–º —Å–µ–º–∏–Ω–∞—Ä–µ –∏ –≤ —Ü–µ–ª–æ–º –≤ –∫—É—Ä—Å–µ –±—É–¥–µ–º –æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å—Å—è –Ω–∞ MSE.

## –õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è
### –ú–∞—Ç—Ä–∏—Ü–∞ –æ–±—ä–µ–∫—Ç—ã-–ø—Ä–∏–∑–Ω–∞–∫–∏

–î–∞–Ω–Ω—ã–π —Å–µ–º–∏–Ω–∞—Ä –ø–æ—Å–≤—è—â—ë–Ω –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ - –≤–∞–∂–Ω–æ–º—É —Ä–∞–∑–¥–µ–ª—É –∑–∞–¥–∞—á–∏ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏, –ø–æ—Ç–æ–º—É —á—Ç–æ –≤ –Ω—ë–º –ª–µ–≥–∫–æ –ø–æ–ª—É—á–∞—é—Ç—Å—è –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.

–í –æ–±—â–µ–º —Å–ª—É—á–∞–µ, –ª–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è - —ç—Ç–æ –ø—Ä–æ—Å—Ç–æ –ø–æ–∏—Å–∫ –Ω–∞–∏–ª—É—á—à–µ–π –∞–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏–∏ —Å—Ä–µ–¥–∏ –º–æ–¥–µ–ª–µ–π –≤–∏–¥–∞

$$
\check y (x) = g_\mathbf{w}(x) = \sum^m_{k=1} \text{w}_k \cdot f_k (x)
$$

–≥–¥–µ $\{f_k(x)\}^m_{k=1}$ - –∫–∞–∫–æ–π-—Ç–æ –Ω–∞–±–æ—Ä –∏–∑ $m$ —Ñ—É–Ω–∫—Ü–∏–π (–∫—Ä–∞–π–Ω–µ –∂–µ–ª–∞—Ç–µ–ª—å–Ω–æ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö).

–í –∏–¥–µ–∞–ª–µ (–¥–ª—è –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ –ª—é–±–æ–≥–æ loss-a) , –º—ã –¥–æ–ª–∂–Ω—ã –ø–æ–ª—É—á–∏—Ç—å –ø–æ–ª–Ω–æ–µ —Å–æ–≤–ø–∞–¥–µ–Ω–∏–µ –Ω–∞ –ª—ç–π–±–ª–∞—Ö (–ø–∞—Ä–∞—Ö $(x_i, y_i))$:

$$
\forall x_i \; \rightarrow \; \check y (x_i) = \sum^m_{k=1} \text{w}_k \cdot f_k (x_i) \stackrel{\text{want}}{=} y_i
$$

–ü–µ—Ä–µ–ø–∏—Å—ã–≤–∞—è –≤ –º–∞—Ç—Ä–∏—á–Ω–æ–º –≤–∏–¥–µ:

$$
\left(\begin{array}{c}
\check y_1 \\
\check y_2 \\
\ldots \\
\check y_N
\end{array}\right)=
\left(\begin{array}{cccc}
f_1\left(x_1\right) & f_2\left(x_1\right) & \ldots & f_m\left(x_1\right) \\
f_1\left(x_2\right) & f_2\left(x_2\right) & \ldots & f_m\left(x_2\right) \\
\ldots & \ldots & \ldots & \ldots \\
f_1\left(x_N\right) & f_2\left(x_N\right) & \ldots & f_m\left(x_N\right)
\end{array}\right)\left(\begin{array}{c}
\text{w}_1 \\
\text{w}_2 \\
\ldots \\
\text{w}_N
\end{array}\right)\stackrel{\text{want}}{=}
\left(\begin{array}{c}
y_1 \\
y_2 \\
\ldots \\
y_N
\end{array}\right)
$$

–ò–ª–∏ –∫–æ—Ä–æ—á–µ:

$$
\check Y = X \cdot \mathbf{w} \stackrel{\text{want}}{=} Y
$$

>__–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ.__ –ù–∞–∑–æ–≤—ë–º —Å—Ç—Ä–æ–∫–∏ –º–∞—Ç—Ä–∏—Ü—ã $X$ - __–æ–±—ä–µ–∫—Ç–∞–º–∏__, –∞ —Å—Ç–æ–ª–±—Ü—ã - __–ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏__. –õ–µ–≥–∫–æ –ø—Ä–∏–Ω—è—Ç—å, —á—Ç–æ —Ç–æ–≥–¥–∞ –º–∞—Ç—Ä–∏—Ü—É $X$ –Ω–∞–∑–æ–≤—ë–º __–º–∞—Ç—Ä–∏—Ü–µ–π –æ–±—ä–µ–∫—Ç—ã-–ø—Ä–∏–∑–Ω–∞–∫–∏__.
>
> –¢–æ –µ—Å—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤–æ –≤—Å–µ—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è $x_i$ - $f_1(x_i), \ldots, f_m(x_i)$ - —ç—Ç–æ –æ–±—ä–µ–∫—Ç.
>
>–î–∞–ª–µ–µ –±—É–¥–µ–º —á–∞—Å—Ç–æ —Ä–∞–±–æ—Ç–∞—Ç—å —Å –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞–º–∏ —ç—Ç–æ–π –º–∞—Ç—Ä–∏—Ü—ã (—Ç–µ–Ω–∑–æ—Ä–Ω—ã–µ –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è) $X = (x_{ij}) = (f_j(x_i))$, –≥–¥–µ $i$ - –Ω–æ–º–µ—Ä —Å—Ç—Ä–æ–∫–∏, $j$ - –Ω–æ–º–µ—Ä —Å—Ç–æ–ª–±—Ü–∞.

__–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ.__ –°–∏–º–≤–æ–ª $\stackrel{\text{want}}{=}$ –∑–¥–µ—Å—å –Ω–µ –æ–∑–Ω–∞—á–∞–µ—Ç —Å—Ç—Ä–æ–≥–æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–∞ (–µ–≥–æ –º–æ–∂–µ—Ç –Ω–µ –±—ã—Ç—å –∫–∞–∫–∏–µ –±—ã –Ω–∏ –±—Ä–∞–ª–∏  $\mathbf{w}$). –ü–æ–Ω–∏–º–∞—Ç—å –µ–≥–æ —Å—Ç–æ–∏—Ç —Å —Ç–æ—á–∫–∏ —Å—Ä–µ–∑–Ω–∏—è –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–≥–æ loss-–∞ (–≤ —á–∞—Å—Ç–Ω–æ—Å—Ç–∏, –¥–∞–ª–µ–µ –ø–æ–ª—É—á–∏–º –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç –¥–ª—è MSE).

### –¢–æ—á–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –¥–ª—è MSE

–ù–∞—á–Ω—ë–º —Å –∞–Ω–µ–∫–¥–æ—Ç–∞.

> __–í–Ω–∏–º–∞–Ω–∏–µ!__ –†–∞—Å—Å—É–∂–¥–µ–Ω–∏—è –Ω–∏–∂–µ –Ω–µ –Ω–∞–¥–æ –≤–æ—Å–ø—Ä–∏–Ω–∏–º–∞—Ç—å —Å–µ—Ä—å–µ–∑–Ω–æ.

–•–æ—Ç–∏–º –Ω–∞–π—Ç–∏ —Ç–∞–∫–æ–π $\mathbf{w}^*$, —á—Ç–æ–±—ã $X \cdot \mathbf{w}^* = Y$. –î–æ–º–Ω–æ–∂–∏–º —Å–ª–µ–≤–∞ –Ω–∞ $X^\top$, —á—Ç–æ–±—ã —Å–¥–µ–ª–∞—Ç—å –º–∞—Ç—Ä–∏—Ü—É –≤ –ª–µ–≤–æ–π —á–∞—Å—Ç–∏ —É—Ä–∞–≤–Ω–µ–Ω–∏—è –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–π:

$$
X^\top \cdot X \cdot \mathbf{w}^* = X^\top Y
$$

–ù–∞–¥–µ–µ–º—Å—è\–≤–µ—Ä–∏–º\–ø—Ä–µ–¥–ø–æ–ª–∞–≥–∞–µ–º, —á—Ç–æ $(X^\top \cdot X)$  –æ–±—Ä–∞—Ç–∏–º–∞, –∞ —Ç–æ–≥–¥–∞ –º–æ–∂–µ–º –¥–æ–º–Ω–æ–∂–∏—Ç—å –Ω–∞ –æ–±—Ä–∞—Ç–Ω—É—é —Å–ª–µ–≤–∞

$$
\mathbf{w}^* = \left( X^\top \cdot X \right)^{-1} \cdot X^\top \cdot Y
$$

–í—É–∞–ª—è, –Ω–∞—à–ª–∏ –∏—Å–∫–æ–º—ã–π —Å—Ç–æ–ª–±–µ—Ü –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –∏ —Ä–ï—à–ò–ª–ò –∑–∞–¥–∞—á—É —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –≤ –æ–±—â–µ–º —Å–ª—É—á–∞–µ –¥–ª—è –ª–Æ–±–û–≥–û loss-–∞ –∏ –ª–Æ–±–û–≥–û –Ω–∞–±–æ—Ä–∞ —Ñ—É–Ω–∫—Ü–∏–π $\{f_k(x)\}^m_{k=1}$, —Ç.–∫. —Ä–∞–≤–µ–Ω—Å—Ç–≤–æ –¥–æ—Å—Ç–∏–≥–Ω—É—Ç–æ üòèüòèüòè

> __–ö–æ–Ω–µ—Ü –≤–Ω–∏–º–∞–Ω–∏—è!__ –†–∞—Å—Å—É–∂–¥–µ–Ω–∏—è –Ω–∏–∂–µ –Ω—É–∂–Ω–æ –≤–æ—Å–ø—Ä–∏–Ω–∏–º–∞—Ç—å —Å–µ—Ä—å–µ–∑–Ω–æ.

–ù–∞ —Å–∞–º–æ–º –¥–µ–ª–µ –≤—ã—à–µ –º—ã –Ω–µ —Ä–µ—à–∏–ª–∏ –∑–∞–¥–∞—á—É, –∏ –¥–µ–ª–æ –Ω–µ –≤ –æ–±—Ä–∞—Ç–∏–º–æ—Å—Ç–∏ –º–∞—Ç—Ä–∏—Ü—ã. –î–æ–º–Ω–æ–∂–µ–Ω–∏–µ —Å–ª–µ–≤–∞ –Ω–∞ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É - –Ω–∏ —Ä–∞–∑—É –Ω–µ —Ä–∞–≤–Ω–æ—Å–∏–ª—å–Ω—ã–π –ø–µ—Ä–µ—Ö–æ–¥, —Ç–∞–∫ —á—Ç–æ —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏ –º—ã –ø—Ä–æ—Å—Ç–æ –ø–æ–∫–∏–¥–∞–ª–∏—Å—å —Å–∏–º–≤–æ–ª–∞–º–∏ –∏ –ø–æ–ª—É—á–∏–ª–∏ —Å–ª—É—á–∞–π–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ. –¢–µ–º —Å–º–µ—à–Ω–µ–µ, —á—Ç–æ —ç—Ç–æ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –æ—Ç–≤–µ—Ç –≤ —Å–ª—É—á–∞–µ MSE Loss. –î–æ–∫–∞–∂–µ–º —ç—Ç–æ –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ.

__–¢–µ–æ—Ä–µ–º–∞.__ –í —Å–ª—É—á–∞–µ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ c MSE —Ñ—É–Ω–∫—Ü–∏–∏ –æ—à–∏–±–æ–∫, –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –Ω–∞–±–æ—Ä–∞ $\{f_k(x)\}^m_{k=1}$, –Ω–∞–∏–ª—É—á—à–∏–º –Ω–∞–±–æ—Ä–æ–º –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –±—É–¥–µ—Ç –≤–µ–∫—Ç–æ—Ä:

$$
\mathbf{w}^* = \left( X^\top \cdot X \right)^{-1} \cdot X^\top \cdot Y
$$

–î–æ–∫–∞–∑–∞—Ç–µ–ª—å—Å—Ç–≤–æ:

–†–∞—Å–ø–∏—à–µ–º MSE (–ø–æ–¥ –∫–æ–Ω–µ—Ü –ø—Ä–∏–≤–µ–¥—ë–º –≤—Å—ë –∫ —Ç–µ–Ω–∑–æ—Ä–Ω—ã–º –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è–º –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞ –¥–∏—Ñ—Ñ–µ—Ä–µ–Ω—Ü–∏–∏—Ä–æ–≤–∞–Ω–∏—è):

$$
\begin{aligned}
\mathcal{L} & =\frac{1}{N} \sum_{i=1}^N\left(\check{y}_i-y_i\right)^2=\left\{\check{y}_i=x_{i j} \text{w}_i\right\}=\frac{1}{N}(X \cdot \mathbf{w}-Y)^{\top}(X \cdot \mathbf{w}-Y)= \\
& =\frac{1}{N}\left(\mathbf{w}^{\top} X^{\top} X \mathbf{w}-Y^{\top} X \cdot \mathbf{w}-\mathbf{w}^{\top} \cdot X^{\top} \cdot Y+Y^{\top} Y\right)= \\
& =\frac{1}{N}\left(\text{w}_j x_{i j} x_{i s} \text{w}_s-y_i \cdot x_{i j} \cdot \text{w}_j-x_{i j} \text{w}_j \cdot y_i+y_i y_i\right)
\end{aligned}
$$

–î–ª—è –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏—è –º–∏–Ω–∏–º—É–º–∞ —Ñ—É–Ω–∫—Ü–∏–∏ –º–Ω–æ–≥–∏—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö –Ω–∞–¥–æ –Ω–∞–π—Ç–∏ –Ω—É–ª–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–Ω–æ–π:

$$
\begin{aligned}
\frac{\partial \mathcal{L}}{\partial \text{w}_k} & =\left\{\begin{array}{l}
\frac{\partial \text{w}_i}{\partial \text{w}_k}=\delta_{i k} \\
A_{k s} \delta_{s m}=A_{k m}
\end{array}\right\}=\frac{1}{N}\left(\delta_{j k} x_{i j} x_{i s} \text{w}_s+\text{w}_j x_{i j} x_{i s} \delta_{s k}-y_i x_{i j} \delta_{j k}-x_{i j} \delta_{j k} \cdot y_i+0\right)= \\
& =\frac{2}{N}\left(x_{i k} x_{i s} \text{w}_s-x_{i k} y_i\right)=0
\end{aligned}
$$

–ê —Ç–æ–≥–¥–∞ –ø–æ–ª—É—á–∞–µ–º —É–∂–µ –∑–Ω–∞–∫–æ–º–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ (–≤ —ç—Ç–æ—Ç —Ä–∞–∑ –ø–µ—Ä–µ—Ö–æ–¥ –∫ –Ω–µ–º—É –∞–±—Å–æ–ª—é—Ç–Ω–æ "–∑–∞–∫–æ–Ω–Ω—ã–π"):

$$
X^{\top} X \cdot \mathbf w-X^{\top} Y=0
$$

–û–ø—è—Ç—å-—Ç–∞–∫–∏ –≤—ã–Ω—É–∂–¥–µ–Ω—ã –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å –æ–±—Ä–∞—Ç–∏–º–æ—Å—Ç—å $(X^\top \cdot X)$, –∏—Ç–æ–≥–æ,

$$
\mathbf w ^ *=\left(X^{\top} X\right)^{-1} X^{\top} Y
$$

> __–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ.__ –î–ª—è –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã $X$ (–¥–∞–∂–µ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–æ–π) –Ω–∞–∑–æ–≤—ë–º __–ø—Å–µ–≤–¥–æ–æ–±—Ä–∞—Ç–Ω–æ–π__ –º–∞—Ç—Ä–∏—Ü—É $\left(X^{\top} X\right)^{-1} X^{\top}$


- __–ü—Ä–∏–º–µ—Ä —Å –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏__

–ü—É—Å—Ç—å –∏–º–µ–µ–º –Ω–∞–±–æ—Ä $\left\{\left(\mathbf{x}_i, y_i\right)\right\}_{i=1}^N$, –≥–¥–µ $\mathbf{x}_i \in \mathbb{R}^3$.

–ü—É—Å—Ç—å —ç—Ç–æ—Ç –Ω–∞–±–æ—Ä –æ–ø—Ä–µ–¥–µ–ª—ë–Ω —Ñ—É–Ω–∫—Ü–∏–µ–π:

$$
y \left(x^1, x^2, x^3 \right) = \mathbf{3} \cdot x^1 - \mathbf{2} \cdot x^2 + \mathbf{5} \cdot x^3 - \mathbf{4}
$$

(–°–¥–µ–ª–∞–µ–º –≤–∏–¥, —á—Ç–æ –º—ã –Ω–µ –∑–Ω–∞–µ–º —ç—Ç–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –∏ –∫–∞–∫ —Ä–∞–∑-—Ç–∞–∫–∏ –∏—Ö –Ω–∞–¥–æ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å) –ó–¥–µ—Å—å –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏–µ $x^k$ - –Ω–µ —Å—Ç–µ–ø–µ–Ω—å, –∞ –Ω–æ–º–µ—Ä –æ—Å–∏.

–¢–æ –µ—Å—Ç—å —Ö–æ—Ç–∏–º –∑–∞—Ñ–∏—Ç–∏—Ç—å –∫–∞–∫—É—é-—Ç–æ —Ñ—É–Ω–∫—Ü–∏—é —Ç—Ä—ë—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö. –í–ø–æ–ª–Ω–µ —Å–µ–±–µ –∑–∞–∫–æ–Ω–Ω–æ –≤—ã–±—Ä–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏–∏ $f_1, f_2, f_3$ –∫–∞–∫ –ø—Ä–æ–µ–∫—Ü–∏–∏ –≤–µ–∫—Ç–æ—Ä–∞ $\mathbf{x}$ –Ω–∞ —Ç—Ä–∏ –æ—Å–∏ –∏–∑ $\mathbb{R}^3$ - —ç—Ç–æ –±—É–¥—É—Ç –Ω–∞—à–∏ –ø—Ä–∏–∑–Ω–∞–∫–∏. –¢–µ–º –Ω–µ –º–µ–Ω–µ–µ –¥–∞–∂–µ –≤–∏–¥–∏–º, —á—Ç–æ –∏—Ö –Ω–µ —Ö–≤–∞—Ç–∏—Ç, –∏ –Ω–∞–¥–æ –¥–æ–±–∞–≤–∏—Ç—å $f_4(x) = 1$ –∫–∞–∫ —á–µ—Ç–≤—ë—Ä—Ç—ã–π –ø—Ä–∏–∑–Ω–∞–∫ (__intercept__ –∏–ª–∏ __bias__).

–ò—Ç–æ–≥–æ, –º–∞—Ç—Ä–∏—Ü–∞ –æ–±—ä–µ–∫—Ç—ã-–ø—Ä–∏–∑–Ω–∞–∫–∏ –±—É–¥–µ—Ç –≤—ã–≥–ª—è–¥–µ—Ç—å –∫–∞–∫

$$
X=\left(\begin{array}{cccc}
x^1_1 & x^2_1 & x^3_1 & 1 \\
x^1_2 & x^2_2 & x^3_2 & 1 \\
\ldots & \ldots & \ldots & \ldots \\
x^1_N & x^2_N & x^3_N & 1
\end{array}\right)
$$

–û–∂–∏–¥–∞–µ–º, —á—Ç–æ —Å–º–æ–∂–µ–º –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –≤—Å–µ –∫–æ—ç—ç—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –∏–∑ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π —Ä–∞–Ω–µ–µ —Ñ–æ—Ä–º—É–ª—ã:

$$
\mathbf w ^ *=\left(X^{\top} X\right)^{-1} X^{\top} Y
$$

```python
import numpy as np

# –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫—É –¥–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è

x1 = np.random.random(100).reshape((-1, 1))
x2 = np.random.random(100).reshape((-1, 1))
x3 = np.random.random(100).reshape((-1, 1))

y_label = 3*x1 - 2*x2 + 5*x3 - 4 # –ù–∞—à–∞ —Ü–µ–ª—å –¥–∞–ª—å—à–µ - –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å —ç—Ç–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã

# –°–æ–∑–¥–∞—ë–º –º–∞—Ç—Ä–∏—Ü—É –æ–±—ä–µ–∫—Ç—ã-–ø—Ä–∏–∑–Ω–∞–∫–∏
X = np.concatenate([x1, x2, x3, np.ones(x1.shape)], axis = 1)

# –°—á–∏—Ç–∞–µ–º –ø—Å–µ–≤–¥–æ–æ–±—Ä–∞—Ç–Ω—É—é –∏ —Å—Ä–∞–∑—É –∂–µ —É–º–Ω–æ–∂–∞–µ–º –Ω–∞ y_label –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤
w = np.linalg.pinv(X) @ y_label # –õ—É—á—à–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å pinv, —á–µ–º —Å–∞–º–æ–º—É –ø–∏—Å–∞—Ç—å –≤—ã—Ä–∞–∂–µ–Ω–∏–µ
w                               # –¥–ª—è –ø–æ—á—Ç–∏ –≤—ã—Ä–æ–∂–¥–µ–Ω–Ω—ã—Ö –º–∞—Ç—Ä–∏—Ü —Å—Ä–∞–±–æ—Ç–∞–µ—Ç –ª—É—á—à–µ
```

- __–ü—Ä–∏–º–µ—Ä —Å –º–Ω–æ–≥–æ—á–ª–µ–Ω–∞–º–∏__

–ü—É—Å—Ç—å –∏–º–µ–µ–º –Ω–∞–±–æ—Ä $\left\{\left(x_i, y_i\right)\right\}_{i=1}^N$, –≥–¥–µ $x_i \in \mathbb{R}$.

–ü—É—Å—Ç—å —ç—Ç–æ—Ç –Ω–∞–±–æ—Ä –æ–ø—Ä–µ–¥–µ–ª—ë–Ω —Ñ—É–Ω–∫—Ü–∏–µ–π:

$$
y \left(x \right) = - \mathbf{4} + \mathbf{3} \cdot x^1 - \mathbf{2} \cdot x^2 + \mathbf{5} \cdot x^3
$$

(–°–¥–µ–ª–∞–µ–º –≤–∏–¥, —á—Ç–æ –º—ã –Ω–µ –∑–Ω–∞–µ–º —ç—Ç–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –∏ –∫–∞–∫ —Ä–∞–∑-—Ç–∞–∫–∏ –∏—Ö –Ω–∞–¥–æ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å) –ê –∑–¥–µ—Å—å –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏–µ $x^k$ - —Å—Ç–µ–ø–µ–Ω—å, –∞ __–Ω–µ__ –Ω–æ–º–µ—Ä –æ—Å–∏.

–ú—ã –æ–∂–∏–¥–∞–µ–º, —á—Ç–æ –∑–∞—Ñ–∏—Ç–∏–º —ç—Ç—É –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –∫–∞–∫–∏–º-—Ç–æ –º–Ω–æ–≥–æ—á–ª–µ–Ω–æ–º, –Ω–æ –∑–∞—Ä–∞–Ω–µ–µ –µ–≥–æ —Å—Ç–µ–ø–µ–Ω—å –Ω–µ –∑–Ω–∞–µ–º. –ü–æ—ç—Ç–æ–º—É –Ω–∞—à–∏–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ –±—É–¥—É—Ç –º–Ω–æ–≥–æ—á–ª–µ–Ω—ã –¥–æ —Å—Ç–µ–ø–µ–Ω–∏ $m$ –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ. –û–ø—è—Ç—å-—Ç–∞–∫–∏ –Ω–µ –∑–∞–±—ã–≤–∞–µ–º –¥–æ–±–∞–≤–∏—Ç—å intercept.

–ú–∞—Ç—Ä–∏—Ü–∞ –æ–±—ä–µ–∫—Ç—ã-–ø—Ä–∏–∑–Ω–∞–∫–∏, —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –±—É–¥–µ—Ç:

$$
X=\left(\begin{array}{cc—Åcc}
x^1_1 & x^2_1 & \ldots & x^m_1 & 1 \\
x^1_2 & x^2_2 & \ldots & x^m_2 & 1 \\
\ldots & \ldots & \ldots & \ldots & \ldots \\
x^1_N & x^2_N & \ldots & x^m_N & 1
\end{array}\right)
$$

–û–∂–∏–¥–∞–µ–º, —á—Ç–æ —Å–º–æ–∂–µ–º –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å –≤—Å–µ –∫–æ—ç—ç—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –∏–∑ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π —Ä–∞–Ω–µ–µ —Ñ–æ—Ä–º—É–ª—ã:

$$
\mathbf w ^ *=\left(X^{\top} X\right)^{-1} X^{\top} Y
$$

–ê –ø–∞—Ä–∞–º–µ—Ç—Ä $m$ (—Ç–∞–∫–∏–µ "–≤–Ω–µ—à–Ω–∏–µ" –ø–∞—Ä–º–µ—Ç—Ä—ã –Ω–∞–∑—ã–≤–∞—é—Ç—Å—è __–≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏__ –º–æ–¥–µ–ª–∏, —Ç.–∫. –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ –∏—Ö –º—ã –Ω–µ —Ñ–∏—Ç–∏–º) –±—É–¥–µ–º –æ–ø—Ä–µ–¥–µ–ª—è—Ç—å –ø–æ –≤–µ–ª–∏—á–∏–Ω–µ –æ—à–∏–±–∫–∏ MSE (—Ç–æ, –≥–¥–µ –º–µ–Ω—å—à–µ, —Ç–∞–º –∏ –ª—É—á—à–µ, –∫–∞–∑–∞–ª–æ—Å—å –±—ã üòè, - –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ MSE –æ—Ç $m$).

–î–æ–±–∞–≤–∏–º –Ω–∞ —ç—Ç–æ—Ç —Ä–∞–∑ –Ω–µ–±–æ–ª—å—à–æ–π —à—É–º –≤ –∏—Å—Å–ª–µ–¥—É–µ–º—É—é –≤—ã–±–æ—Ä–∫—É –¥–ª—è —Å–ª–æ–∂–Ω–æ—Å—Ç–∏.

```python
import numpy as np

# –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫—É –¥–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è

x = 2*np.random.random(200).reshape((-1, 1)) - 1
noise = 0.1*np.random.normal(size = x.shape)
y_label = -4 + 3*x - 2*x**2 + 5*x**3 + noise

# –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è —Å–æ–∑–¥–∞–Ω–∏—è –º–∞—Ç—Ä–∏—Ü—ã –æ–±—ä–µ–∫—Ç—ã-–ø—Ä–∏–∑–Ω–∞–∫–∏
def create_X(x, m):
    X = np.ones(x.shape) # –°—Ä–∞–∑—É –∂–µ –¥–æ–±–∞–≤–∏–º intercept
    for i in range(1, m + 1):
        X = np.concatenate([X, x**i], axis = 1)
    return X


# –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Å–æ–≤ –ø—Ä–∏ –∑–∞–¥–∞–Ω–Ω–æ–º m
def fit(x, y_label, m):
    X = create_X(x, m)
    w = np.linalg.pinv(X) @ y_label
    return w

# –§—É–Ω–∫—Ü–∏—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è y_predict –¥–ª—è –ª—é–±–æ–≥–æ x –º–æ–¥–µ–ª–∏ —Å –≤–µ—Å–∞–º–∏ w
def predict(x, w):
    return create_X(x, np.size(w) - 1) @ w

# –§—É–Ω–∫—Ü–∏—è –æ—à–∏–±–∫–∏ MSE –¥–ª—è –º–æ–¥–µ–ª–∏ (–∫–æ—Ç–æ—Ä–∞—è –ø–æ–ª–Ω–æ—Å—Ç—å—é –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è–º–∏ –º–æ–¥–µ–ª–∏ –∏ y_label)
def MSE(y, y_label):
    return np.sum((y - y_label)**2) / np.size(y)


fit(x, y_label, 3)
```

```python
# –ü–æ—Å—Ç—Ä–æ–∏–º MSE –æ—Ç m
for m in range(10):
    w = fit(x, y_label, m)
    y_predict = predict(x, w)
    print(f"for {m} MSE is {MSE(y_predict, y_label)}")
```

–û—à–∏–±–∫–∞ —É–º–µ–Ω—å—à–∞–µ—Ç—Å—è —Å —Ä–æ—Å—Ç–æ–º m, –Ω–æ –∫–∞–∫ –∂–µ —Ç–∞–∫, —É –Ω–∞—Å –∂–µ –Ω–µ –º–Ω–æ–≥–æ—á–ª–µ–Ω –±–æ–ª—å—à–æ–π —Å—Ç–µ–ø–µ–Ω–∏?

–î–∞ –ø—Ä–æ—Å—Ç–æ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ø—Ä–∏ –±–æ–ª—å—à–∏—Ö —Å—Ç–µ–ø–µ–Ω—è—Ö –±—É–¥—É—Ç _–º–∞–ª–µ–Ω—å–∫–∏–µ_:

```python
fit(x, y_label, 5) # –ó–Ω–∞—á–∏—Ç –ª–∏ —ç—Ç–æ, —á—Ç–æ –º–æ–∂–µ–º —Å—Ç–∞–≤–∏—Ç—å m —Å–∫–æ–ª—å —É–≥–æ–¥–Ω–æ –±–æ–ª—å—à–∏–º? –î–∞ –∏ –Ω–µ —Ç–æ —á—Ç–æ–±—ã –æ–Ω–∏ —Å–æ–≤—Å–µ–º —É–∂–µ –º–∞–ª–µ–Ω—å–∫–∏–µ
```

## –ü–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ

```python
import numpy as np
import matplotlib.pyplot as plt

# –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º –≤—ã–±–æ—Ä–∫—É –¥–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è
x = 2*np.random.random(100).reshape((-1, 1)) - 1
noise = 0.1*np.random.normal(size = x.shape)
y_label = -4 + 3*x - 2*x**2 + 5*x**3 + noise



def plot_graph(m):

    plt.scatter(x, y_label, c='red')

    x_plot = np.linspace(-1, 1, 10000).reshape((-1, 1))

    w = fit(x, y_label, m)

    y_plot = create_X(x_plot, m) @ w

    plt.plot(x_plot, y_plot)

    plt.ylim(-14, 0)

    error = MSE(y_label, create_X(x, m) @ w)
    plt.text(0.5, -10, f"MSE is {error:.5f}")



m_slider = widgets.IntSlider(value=3, min=0, max=100, step=1, description='m')

interact(plot_graph, m = m_slider)
```

–í–∏–¥–∏–º, —á—Ç–æ –æ—à–∏–±–∫–∞ –ø—Ä–æ–¥–æ–ª–∂–∞–µ—Ç —É–º–µ–Ω—å—à–∞—Ç—å—Å—è —Å —Ä–æ—Å—Ç–æ–º $m$, —Ö–æ—Ç—è —Å —Ñ—É–Ω–∫—Ü–∏–µ–π —è–≤–Ω–æ –ø—Ä–æ–∏—Å—Ö–æ–¥–∏—Ç —á—Ç–æ-—Ç–æ —Å—Ç—Ä–∞—à–Ω–æ–µ –∏ –æ–Ω–∞ —É–∂–µ —è–≤–Ω–æ –Ω–µ –æ–ø–∏—Å—ã–≤–∞–µ—Ç –∏—Å—Ö–æ–¥–Ω—ã–π –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö.

–¢–∞–∫–∞—è —Å–∏—Ç—É–∞—Ü–∏—è –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è __–ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ–º__ - –∫–æ–≥–¥–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å–ª–∏—à–∫–æ–º –º–Ω–æ–≥–æ –∏ –º–æ–¥–µ–ª—å —Å–≤–æ–µ–≥–æ —Ä–æ–¥–∞ "–≤—ã—É—á–∏–≤–∞–µ—Ç" –æ–±—É—á–∞—é—â—É—é –≤—ã–±–æ—Ä–∫—É, –ø—Ä–∏ —ç—Ç–æ–º —Ç–µ—Ä—è–µ—Ç –≤—Å—è–∫—É—é –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç–µ–ª—å–Ω—É—é —Å–ø–æ—Å–æ–±–Ω–æ—Å—Ç—å –∑–∞ –µ—ë –ø—Ä–µ–¥–µ–ª–∞–º–∏.

–ú—ã –≤–ø–æ–ª–Ω–µ –º–æ–≥–ª–∏ –∑–∞—Ä–∞–Ω–µ–µ –æ—Ç–ª–æ–≤–∏—Ç—å —ç—Ç—É —Å–∏—Ç—É–∞—Ü–∏—é, –µ—Å–ª–∏ –±—ã –∏–∑–º–µ—Ä—è–ª–∏ MSE –Ω–µ –Ω–∞ __–æ–±—É—á–∞—é—â–µ–π__ –≤—ã–±–æ—Ä–∫–µ (—Ç–æ–π, –Ω–∞ –∫–æ—Ç–æ—Ä–æ–π —Ñ–∏—Ç–∏–ª–∏), –∞ –Ω–∞ –Ω–µ–∫–æ—Ç–æ—Ä–æ–π __—Ç–µ—Å—Ç–æ–≤–æ–π__.

## –ú–µ—Ç–æ–¥ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç (PCA)
### –ö–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞

–†–∞—Å—Å–º–æ—Ç—Ä–∏–º –ø–æ–¥—Ä–æ–±–Ω–µ–µ –º–∞—Ç—Ä–∏—Ü—É –æ–±—ä–µ–∫—Ç—ã-–ø—Ä–∏–∑–Ω–∞–∫–∏.

$$
X=
\left(\begin{array}{cccc}
f_1\left(x_1\right) & f_2\left(x_1\right) & \ldots & f_m\left(x_1\right) \\
f_1\left(x_2\right) & f_2\left(x_2\right) & \ldots & f_m\left(x_2\right) \\
\ldots & \ldots & \ldots & \ldots \\
f_1\left(x_N\right) & f_2\left(x_N\right) & \ldots & f_m\left(x_N\right)
\end{array}\right)
$$

–ê —Ç–æ—á–Ω–µ–µ —Å–ª–µ–¥—É—é—â–µ–µ –µ—ë –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ:

$$
\left(X^\top X\right)_{ij}=\sum^m_{k=1} f_i(x_k) \cdot f_j(x_k)
$$

–í–∏–¥–∏–º, —á—Ç–æ —ç—Ç–æ –±—É–∫–≤–∞–ª—å–Ω–æ –≤—ã–≥–ª—è–¥–∏—Ç –∫–∞–∫ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏—è! –¢–æ–ª—å–∫–æ –≤ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–∏ –Ω–∞–¥–æ, —á—Ç–æ–±—ã —Å–ª—É—á–∞–π–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã (–∞ –Ω–∞—à–∏ –ø—Ä–∏–∑–Ω–∞–∫–∏ - —ç—Ç–æ –ø–æ —Å—É—Ç–∏ —Å–ª—É—á–∞–π–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã) –∏–º–µ–ª–∏ –Ω—É–ª–µ–≤–æ–µ —Å—Ä–µ–¥–Ω–µ–µ. –¢–∞–∫ –¥–∞–≤–∞–π—Ç–µ —Ç–æ–≥–¥–∞ –ø–µ—Ä–µ–π–¥—ë–º –∫ —Ç–∞–∫–∏–º –ø—Ä–∏–∑–Ω–∞–∫–∞–º, —á—Ç–æ–±—ã –∏—Ö —Å—Ä–µ–¥–Ω–µ–µ –Ω–∞ –≤—ã–±–æ—Ä–∫–µ –±—ã–ª–æ –Ω–æ–ª—å.

> __–í–Ω–∏–º–∞–Ω–∏–µ!__ –ó–¥–µ—Å—å –∏ –¥–∞–ª–µ–µ –ø–æ–¥—Ä–∞–∑—É–º–µ–≤–∞–µ–º, —á—Ç–æ —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–æ–≤ –≤ –º–∞—Ç—Ä–∏—Ü–µ –• —Ä–∞–≤–Ω—ã –Ω—É–ª—é. –°—Ç–æ–ª–±–µ—Ü Intercept, –µ—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω–æ, –Ω–∞–¥–æ –±—É–¥–µ—Ç –æ—Ç–±—Ä–æ—Å–∏—Ç—å —Å–æ–≤—Å–µ–º - –¥–ª—è —ç—Ç–æ–≥–æ —Ç–∞–∫–∂–µ –≤—ã—á—Ç–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∏–∑ —Å—Ç–æ–ª–±—Ü–∞ –ª—ç–π–±–ª–æ–≤ $y_i$. –¢–∞–∫–∞—è —à—Ç—É–∫–∞ –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è __—Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ–º__ –¥–∞–Ω–Ω—ã—Ö:
>
>$$
f_k(x) \stackrel{\text{–∑–∞–º–µ–Ω–∞}}{\longmapsto} f_k(x) - \frac{1}{N}\sum^N_{i=1} f_k(x_i)
$$
>$$
y_i \stackrel{\text{–∑–∞–º–µ–Ω–∞}}{\longmapsto} y_i - \frac{1}{N}\sum^N_{i=1} y_i
$$

–¢–æ–≥–¥–∞ –≤ —á–∏—Å—Ç–æ–º –≤–∏–¥–µ, –≤–∏–¥–∏–º __–∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É__

$$
\left(X^\top X\right)_{ij} = \sum^m_{k=1} f_i(x_k) \cdot f_j(x_k) = \text{cov}(f_i, f_j)
$$

–ê —Ç–æ–≥–¥–∞, –∏–∑ —Å–≤–æ–π—Å—Ç–≤ –º–∞—Ç—Ä–∏—Ü—ã $X^\top X$ –ø–æ–Ω–∏–º–∞–µ–º, —á—Ç–æ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Ç–∞–∫–æ–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, —á—Ç–æ –≤ –Ω–æ–≤—ã—Ö "–∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞—Ö" –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –∏–º–µ–µ—Ç –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω—ã–π –≤–∏–¥ (—ç—Ç–æ —Å–ø–µ–∫—Ç—Ä–∞–ª—å–Ω–æ–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ $X^\top X$):

$$
X^\top X = U \cdot \Lambda \cdot U^\top, \quad U - \text{—É–Ω–∏—Ç–∞—Ä–Ω–∞—è}
$$

–¢–æ –µ—Å—Ç—å, –º—ã __–≤—Å–µ–≥–¥–∞__ –º–æ–∂–µ–º –ø–µ—Ä–µ–π—Ç–∏ –æ—Ç —Ç–µ–∫—É—â–µ–≥–æ –Ω–∞–±–æ—Ä–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∫ –Ω–µ–∫–æ—Ä—Ä–µ–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã–º –¥—Ä—É–≥ –∫ –¥—Ä—É–≥—É –ø—Ä–∏–∑–Ω–∞–∫–∞–º, —á—Ç–æ —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ –ø–æ–≤—ã—Å–∏—Ç –æ–±—É—Å–ª–æ–≤–ª–µ–Ω–Ω–æ—Å—Ç—å –∑–∞–¥–∞—á–∏ (–≤—Å–ø–æ–º–Ω–∏—Ç–µ —Å–º—ã—Å–ª —á–∏—Å–ª–∞ –æ–±—É—Å–ª–æ–≤–ª–µ–Ω–Ω–æ—Å—Ç–∏).

–ë–æ–ª–µ–µ —Ç–æ–≥–æ, —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ —á–∏—Å–ª–∞ $\Lambda$ –º–∞—Ç—Ä–∏—Ü—ã $X^\top X$ (–∏–ª–∏ —Å–∏–Ω–≥—É–ª—è—Ä–Ω—ã–µ —á–∏—Å–ª–∞ $\sqrt{\Lambda}$ –º–∞—Ç—Ä–∏—Ü—ã $X$) - –±—É–¥—É—Ç —Å–∫–æ—Ä—Ä–µ–ª–∏—Ä–æ–≤–∞–Ω—ã —Å __–≤–∞–∂–Ω–æ—Å—Ç—å—é__ —Ç–æ–≥–æ –∏–ª–∏ –∏–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ (–ß–µ–º –±–æ–ª—å—à–µ –¥–∏—Å–ø–µ—Ä—Å–∏—è –ø—Ä–∏–∑–Ω–∞–∫–∞, —Ç–µ–º –æ–Ω "–≤–∞–∂–Ω–µ–µ"). –¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –º—ã –º–æ–∂–µ–º –¥–∞–∂–µ –æ—Ç—Å–µ–∫–∞—Ç—å "–Ω–µ–≤–∞–∂–Ω—ã–µ" –ø—Ä–∏–∑–Ω–∞–∫–∏.

–í —ç—Ç–æ–º –∏ –µ—Å—Ç—å —Å—É—Ç—å PCA.

![](https://i.stack.imgur.com/lNHqt.gif)

### –ú–µ—Ç–æ–¥ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç (PCA - principal component analysis)

- __–ö–∞–∫ —Ä–∞–±–æ—Ç–∞–µ—Ç –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤?__

–ü–æ —Å—É—Ç–∏, –∑–∞–º–µ—Ç–∏–º, —á—Ç–æ:

$$
\underbrace{X}_{(N, m)} \cdot \underbrace{V}_{(m,m)} = \underbrace{Z}_{(N,m)},  \quad V - \text{—É–Ω–∏—Ç–∞—Ä–Ω–∞—è}
$$

–û–ø—Ä–µ–¥–µ–ª—è–µ—Ç –ø–æ–≤–æ—Ä–æ—Ç –≤ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

–ò–∑ –Ω–∞—à–µ–π —Ü–µ–ª–∏ –¥–∏–∞–≥–æ–Ω–∞–ª–∏–∑–∞—Ü–∏–∏ $X^\top X$, –º—ã —Ö–æ—Ç–∏–º –≤–∏–¥–∏–º:

$$
X^\top X = V \cdot Z^\top Z \cdot V^\top = \left|V ‚â° U - \text{—Å–æ–±—Å—Ç–≤. –≤–µ–∫—Ç–æ—Ä–∞ $X^\top X$} \right| = U \cdot \underbrace{Z^\top Z}_{\Lambda} \cdot U^\top
$$

–¢–æ –µ—Å—Ç—å $Z = X \cdot U$ - –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –Ω—É–∂–Ω–æ–µ –Ω–∞–º –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∫ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –æ—Ä—Ç–æ–≥–æ–Ω–∞–ª—å–Ω—ã–º –¥—Ä—É–≥ –¥—Ä—É–≥—É –ø—Ä–∏–∑–Ω–∞–∫–∞–º.

>–ù–∞ –ø—Ä–∞–∫—Ç–∏–∫–µ —ç—Ç–æ –≤—ã–≥–ª—è–¥–∏—Ç –∫–∞–∫
>
>$$X \cdot \mathbf{w} \stackrel{\text{want}}{=} Y$$
>
>$$Z  \cdot U^\top \cdot \mathbf{w} \stackrel{\text{want}}{=} Y$$
>
>$$Z  \cdot \widetilde{\mathbf{w}} \stackrel{\text{want}}{=} Y$$

- __–û–±—Ä–µ–∑–∫–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤__

–í—Å–ø–æ–º–∏–Ω–∞—è —Å–∏–Ω–≥—É–ª—è—Ä–Ω–æ–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ,

$$
X=\underbrace{V}_{(N, m)} \cdot \underbrace{\sqrt{\Lambda}}_{(m, m)} \cdot \underbrace{U^{T}}_{(m, N)}=V\left(\begin{array}{ccc}
\sqrt{\lambda_1} & & \\
& \ddots & &  \\
& & &\ddots & \\
& & & & \sqrt{\lambda_m}
\end{array}\right) U^{T}, \quad \text{–≥–¥–µ } U, V \text{- —É–Ω–∏—Ç–∞—Ä–Ω—ã–µ, –∏ } \lambda_1 \geqslant \lambda_2 \geqslant \lambda_3 \geqslant \ldots \geqslant \lambda_m
$$

–í–æ–∑—å–º—ë–º –∏ –º–æ–¥–∏—Ñ–∏—Ü–∏—Ä—É–µ–º —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É, —É–±—Ä–∞–≤ —Å–∞–º—ã–µ –Ω–∞–∏–º–µ–Ω—å—à–∏–µ —Å–∏–Ω–≥—É–ª—è—Ä–Ω—ã–µ —á–∏—Å–ª–∞, –æ—Å—Ç–∞–≤–∏–≤ $k$ –Ω–∞–∏–±–æ–ª—å—à–∏—Ö. –°–æ—Å—Ç–∞–≤–∏–º —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –Ω–æ–≤—É—é –º–∞—Ç—Ä–∏—Ü—É –æ–±—ä–µ–∫—Ç—ã-–ø—Ä–∏–∑–Ω–∞–∫–∏:

$$
\widetilde{X}=V\left(\begin{array}{cccc}
\sqrt{\lambda_1} & & & \\
& \ddots & & \\
& & \sqrt{\lambda_k} & & \\
& & & \ddots & \\
& & & & 0\\
\end{array}\right) U^{T}=\underbrace{V \cdot \sqrt{\widetilde{\Lambda}}}_{\widetilde{Z}} \cdot U^T=\widetilde{Z} \cdot U^{T}
$$

__–¢–µ–æ—Ä–µ–º–∞:__: –≠—Ç–∞ $\widetilde{Z}$ –Ω–∞–∏–ª—É—á—à–∏–º –æ–±—Ä–∞–∑–æ–º —Ä–µ—à–∞–µ—Ç –∑–∞–¥–∞—á—É –æ –º–∞–ª–æ—Ä–∞–Ω–≥–æ–≤–æ–º –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–∏

$$
\left\|\widetilde{Z} \cdot \widetilde{U}^{T}-X\right\| \longrightarrow \underset{\widetilde{Z}, \widetilde{U}}{\min}
$$

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –æ—Å—Ç–∞—ë—Ç—Å—è –ø—Ä–æ—Å—Ç–æ —Ä–µ—à–∞—Ç—å –∑–∞–¥–∞—á—É —Å –Ω–æ–≤–æ–π –º–∞—Ç—Ä–∏—Ü–µ–π –æ–±—ä–µ–∫—Ç—ã-–ø—Ä–∏–∑–Ω–∞–∫–∏ $\widetilde{Z} = \widetilde{X} \cdot U$.

### –ü—Ä–∏–º–µ—Ä PCA –≤ –±–æ—Ä—å–±–µ —Å –ª–∏—à–Ω–∏–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
# –ü—Ä–∏–º–µ—Ä —Å –æ–±—Ä–µ–∑–∫–æ–π –ª–∏—à–Ω–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

MAX_ORDER = 20 # –í–æ–∑—å–º—ë–º –º–Ω–æ–≥–æ—á–ª–µ–Ω—ã 19 —Å—Ç–µ–ø–µ–Ω–∏ –∏ –Ω–∏–∂–µ (–ø–µ—Ä–≤—ã–µ 20, —Å—á–∏—Ç–∞—è –æ—Ç –Ω—É–ª—è)

xs = 2*np.random.random(200).reshape((-1, 1)) - 1
noise = 0.1*np.random.normal(size = xs.shape)
ys = -4 + 3*xs - 2*xs**2 + 5*xs**3 + noise

xs_train, xs_test, ys_train, ys_test = train_test_split(xs, ys, test_size=0.5, random_state=13)


poly_regresssor = PolynomialFeatures(degree=MAX_ORDER, include_bias=False) # –ê –≤–æ—Ç —Ç–∞–∫ –ø—Ä–æ—Å—Ç–æ –º–æ–∂–Ω–æ —Ñ–∏—Ç–∏—Ç—å –º–Ω–æ–≥–æ—á–ª–µ–Ω–∞–º–∏
X_train = poly_regresssor.fit_transform(xs_train.reshape(-1, 1)) # –ò–ª–∏ –≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å –º–∞—Ç—Ä–∏—Ü—É X
X_test = poly_regresssor.fit_transform(xs_test.reshape(-1, 1))


shift_x = X_train.mean(axis=0).reshape((1, -1)) # –°–¥–≤–∏–≥–∞–µ–º –Ω–∞ –Ω—É–ª–µ–≤–æ–µ —Å—Ä–µ–¥–Ω–µ–µ
X_train = X_train - shift_x
X_test = X_test - shift_x


shift_y = ys_train.mean()
ys_train -= shift_y
ys_test -= shift_y
```

```python
w = np.linalg.pinv(X_train) @ ys_train


y_train_predicted = X_train @ w
print('Train MSE', mean_squared_error(y_train_predicted, ys_train))

y_test_predicted = X_test @ w
print('Test MSE', mean_squared_error(y_test_predicted, ys_test))
```

```python
import seaborn as sns
sns.barplot(x = np.arange(len(w)), y = w.flatten()) # –û—Ç–æ–±—Ä–∞–∑–∏–º –≤–µ—Å–∞ –Ω–∞ –±–∞—Ä–ø–ª–æ—Ç–µ –¥–ª—è —è—Å–Ω–æ—Å—Ç–∏
```

–í–∏–¥–∏–º –æ–≥—Ä–æ–º–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤–µ—Å–æ–≤ - –æ–¥–∏–Ω –∏–∑ —è–≤–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏—è.

–≠—Ç–æ –º–æ–∂–Ω–æ "–Ω–∞ –ø–∞–ª—å—Ü–∞—Ö" –æ–±—ä—è—Å–Ω–∏—Ç—å —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º: –∫–∞–∫–æ–π-—Ç–æ –ø—Ä–∏–∑–Ω–∞–∫ –∏–º–µ–µ—Ç –æ—á–µ–Ω—å –º–∞–ª–µ–Ω—å–∫—É—é –¥–∏—Å–ø–µ—Ä—Å–∏—é, –ø–æ—ç—Ç–æ–º—É —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–π –µ–π –≤–µ—Å –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å –æ—á–µ–Ω—å –±–æ–ª—å—à–∏–º –¥–ª—è "–∏—Å–∫—É—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ" —É—á–∏—Ç—ã–≤–∞–Ω–∏—è —ç—Ç–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞.

–°–¥–µ–ª–∞–µ–º PCA –∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Å –º–∞–ª–µ–Ω—å–∫–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π –º–æ–∂–Ω–æ –æ–±—Ä–µ–∑–∞—Ç—å.

```python
from sklearn.decomposition import PCA # –ó–∞ –Ω–∞—Å —É–∂–µ –≤—Å—ë —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω–æ

pca = PCA(n_components = 20) # –û—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ 10 –Ω–∞–∏–≤–∞–∂–Ω–µ–π—à–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞
pca.fit(X_train)

X_train_reduced = pca.transform(X_train) # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç
X_test_reduced = pca.transform(X_test)



w = np.linalg.pinv(X_train_reduced) @ ys_train

y_train_predicted = X_train_reduced @ w
print('Train MSE', mean_squared_error(y_train_predicted, ys_train))

y_test_predicted = X_test_reduced @ w
print('Test MSE', mean_squared_error(y_test_predicted, ys_test))
```

```python
sing_values = pca.singular_values_

sns.barplot(x = np.arange(len(sing_values)), y = sing_values, log = True)

print("–í–∏–¥–∏–º, —á—Ç–æ 5 –ø—Ä–∏–∑–Ω–∞–∫ –≤ 10 —Ä–∞–∑ –º–µ–Ω—å—à–µ —Å–∞–º–æ–≥–æ –≤–∞–∂–Ω–æ–≥–æ - –º–æ–∂–Ω–æ –∏ –æ–±—Ä–µ–∑–∞—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∏ –¥–∞–ª–µ–µ")
```

### PCA –≤ —Å–∂–∞—Ç–∏–∏ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π

–î–ª—è –ø–ª–æ—Ç–Ω—ã—Ö –º–∞—Ç—Ä–∏—Ü —Ç—Ä–µ–±—É–µ—Ç—Å—è —Ö—Ä–∞–Ω–∏—Ç—å $N^2$ —ç–ª–µ–º–µ–Ω—Ç–æ–≤. –ê–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏—è —Ä–∞–Ω–≥–∞ r (–º–µ—Ç–æ–¥ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç) –º–æ–∂–µ—Ç —É–º–µ–Ω—å—à–∏—Ç—å —ç—Ç–æ —á–∏—Å–ª–æ O($N_r$) . –í –∫–∞—á–µ—Å—Ç–≤–µ –ø—Ä–∏–º–µ—Ä–∞ —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º —Å–∂–∞—Ç–∏–µ –∏–∑–æ–±—Ä–∞–∂–µ–Ω–∏–π.

```python
#!pip install scikit-image

from skimage import data
from skimage.color import rgb2gray
from numpy.linalg import svd
from skimage import img_as_float
```

```python
cat = rgb2gray(img_as_float(data.chelsea()))
cat.shape
```

```python
plt.imshow(cat)
plt.tight_layout()
```

```python
U, s, Vc = svd(cat, full_matrices=False)
def compress_and_show(k):
    os = cat.shape
    cat_compressed = U[:,:k] @ np.diag(s[:k]) @ Vc[:k,:]
    compression_ratio = 100.0* (k*(os[0] + os[1])+k)/(os[0]*os[1])
    plt.title("compression ratio={:.2f}".format(compression_ratio)+"%")
    plt.imshow(cat_compressed)
    plt.axis('off')
    plt.tight_layout()
    return cat_compressed
```

```python
cat_compressed = compress_and_show(100)
cat_compressed.shape
```

```python
from ipywidgets import interact


def on_value_change(b):
    compress_and_show(b)

b_slider = widgets.IntSlider(value=90, min=1, max=100, step=1, description='b')

interact(on_value_change, b=b_slider)
```

## –î–æ–º–∞—à–Ω–µ–µ –∑–∞–¥–∞–Ω–∏–µ


## 1. –ë–∞–∑–æ–≤–∞—è –∑–∞–¥–∞—á–∞ –ú–ù–ö.


**–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–π –≤–æ–ø—Ä–æ—Å 1.**

–ü—É—Å—Ç—å –¥–∞–Ω–∞ –≤—ã–±–æ—Ä–∫–∞ —Ç–æ—á–µ–∫ $y_i$. –†–µ—à–∏—Ç–µ –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–∏ –∑–∞–¥–∞—á—É –ú–ù–ö, –º–æ–¥–µ–ª–∏—Ä—É—è –¥–∞–Ω–Ω—ã–µ –ø–æ—Å—Ç–æ—è–Ω–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω–æ–π $\check{y}$, —á—Ç–æ –æ—Ç–≤–µ—á–∞–µ—Ç –º–∏–Ω–∏–º–∏–∑–∞—Ü–∏–∏ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ—Ç–µ—Ä—å
$$
\mathcal{L}=\sum_{i=1}^l\left(y_i-\check{y}\right)^2 \rightarrow \min _{\breve{y}} .
$$

**–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–π –≤–æ–ø—Ä–æ—Å 2.**

–ü–æ–∫–∞–∂–∏—Ç–µ, —á—Ç–æ –ø—Ä—è–º–∞—è, –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω–∞—è –ø–æ –º–µ—Ç–æ–¥—É –ú–ù–ö, –≤—Å–µ–≥–¥–∞ –ø—Ä–æ—Ö–æ–¥–∏—Ç —á–µ—Ä–µ–∑ —Ç–æ—á–∫—É $(\bar{x}, \bar{y})$, –≥–¥–µ $\bar{x}$ –∏ $\bar{y}$ - –≤—ã–±–æ—Ä–æ—á–Ω—ã–µ —Å—Ä–µ–¥–Ω–∏–µ. –û–±–æ–±—â–∏—Ç–µ –Ω–∞ —Å–ª—É—á–∞–π –º–Ω–æ–≥–æ–º–µ—Ä–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏.

**–ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –∑–∞–¥–∞–Ω–∏–µ.**

–î–ª—è —á–µ—Ç—ã—Ä–µ—Ö –≤—ã–±–æ—Ä–æ–∫ –∏–∑ –∫–≤–∞—Ä—Ç–µ—Ç–∞ –≠–Ω—Å–∫–æ–º–±–∞ –≤—ã—á–∏—Å–ª–∏—Ç–µ –≤—ã–±–æ—Ä–æ—á–Ω—ã–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ $x$ –∏ $y$ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç, –∞ —Ç–∞–∫–∂–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ª–∏–Ω–µ–π–Ω–æ–π –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏ –ü–∏—Ä—Å–æ–Ω–∞. –ò–∑–æ–±—Ä–∞–∑–∏—Ç–µ –≤—ã–±–æ—Ä–∫–∏ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–∞—Ö. –î–∞–Ω–Ω—ã–µ –º–æ–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å –≤ —Å–∏—Å—Ç–µ–º–µ —òu—Ä—Éter —Å –ø–æ–º–æ—â—å—é –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ seaborn, –≤—ã–∑–≤–∞–≤ –º–µ—Ç–æ–¥ `load_dataset('anscombe')` .

## 2. –¶–µ–Ω—Ç—Ä–∞–ª–∏–∑–∞—Ü–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏ –ú–ù–ö.

–ü–æ–∫–∞–∂–∏—Ç–µ, —á—Ç–æ —Å–ª–µ–¥—É—é—â–∏–µ –¥–≤–µ –ø—Ä–æ—Ü–µ–¥—É—Ä—ã –ø—Ä–∏–≤–æ–¥—è—Ç –∫ –æ–¥–∏–Ω–∞–∫–æ–≤–æ–º—É —Ä–µ–∑—É–ª—å—Ç–∞—Ç—É:
1. –í –º–∞—Ç—Ä–∏—Ü–µ –æ–±—ä–µ–∫—Ç-–ø—Ä–∏–∑–Ω–∞–∫ $X$ –∏–∑ –∫–∞–∂–¥–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –≤—ã—á–∏—Ç–∞–µ—Ç—Å—è —Å—Ä–µ–¥–Ω–µ–µ –ø–æ —Å—Ç–æ–ª–±—Ü—É (—Ü–µ–Ω—Ç—Ä–∞–ª–∏–∑–∞—Ü–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤). –ü–æ—Å–ª–µ —ç—Ç–æ–≥–æ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è $\left(X^T X\right)^{-1}$.
2. –ö –º–∞—Ç—Ä–∏—Ü–µ $X$ –¥–æ–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –≤ –∫–æ–Ω–µ—Ü —Å—Ç–æ–ª–±–µ—Ü, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ –æ–¥–Ω–∏—Ö –µ–¥–∏–Ω–∏—Ü. –í—ã—á–∏—Å–ª—è–µ—Ç—Å—è $\left(X^T X\right)^{-1}$ –∏ –≤ –ø–æ–ª—É—á–∏–≤—à–µ–π—Å—è –º–∞—Ç—Ä–∏—Ü–µ –≤—ã—á–µ—Ä–∫–∏–≤–∞–µ—Ç—Å—è –ø–æ—Å–ª–µ–¥–Ω–∏–π —Å—Ç–æ–ª–±–µ—Ü –∏ –ø–æ—Å–ª–µ–¥–Ω—è—è —Å—Ç—Ä–æ–∫–∞.


–¢–∞–∫–∂–µ –ø—Ä–æ–≤–µ—Ä—å—Ç–µ —ç—Ç–æ –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –¥–ª—è —Å–ª—É—á–∞–π–Ω–æ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö –º–∞—Ç—Ä–∏—Ü.

## 3. –ì–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–π —Å–º—ã—Å–ª –ø—Å–µ–≤–¥–æ–æ–±—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã.

–ù–∞ –ª–µ–∫—Ü–∏–∏ –æ–±—Å—É–∂–¥–∞–ª–æ—Å—å, —á—Ç–æ –º–µ—Ç–æ–¥ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ - —ç—Ç–æ —Å–ø–æ—Å–æ–± –ø–æ—Å—Ç–∞–≤–∏—Ç—å –∑–∞–¥–∞—á—É –æ —Ä–µ—à–µ–Ω–∏–∏ –ø–µ—Ä–µ–æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π —Å–∏—Å—Ç–µ–º—ã $X w=y$, –∫–æ—Ç–æ—Ä–∞—è –∏–º–µ–µ—Ç —è–≤–Ω—ã–π –æ—Ç–≤–µ—Ç, –≤—ã—Ä–∞–∂–∞—é—â–∏–π—Å—è —á–µ—Ä–µ–∑ –ª–µ–≤—É—é –ø—Å–µ–≤–¥–æ–æ–±—Ä–∞—Ç–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É –¥–ª—è $X$. –î–ª—è –Ω–µ–¥–æ–æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π —Å–∏—Å—Ç–µ–º—ã $X w=y$ (–∏–º–µ—é—â–µ–π –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ –º–Ω–æ–≥–æ —Ä–µ—à–µ–Ω–∏–π) –º–æ–∂–Ω–æ –ø–æ—Å—Ç–∞–≤–∏—Ç—å –∑–∞–¥–∞—á—É –æ –ø–æ–∏—Å–∫–µ —Ä–µ—à–µ–Ω–∏—è —Å –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–π $l_2$-–Ω–æ—Ä–º–æ–π –≤–µ—Å–æ–≤ $\|w\|^2=w^T w$. –†–µ—à–∏—Ç–µ —Ç–∞–∫—É—é –∑–∞–¥–∞—á—É –∏ –ø–æ–∫–∞–∂–∏—Ç–µ, —á—Ç–æ –æ—Ç–≤–µ—Ç –≤—ã—Ä–∞–∂–∞–µ—Ç—Å—è —á–µ—Ä–µ–∑ –ø—Ä–∞–≤—É—é –ø—Å–µ–≤–¥–æ–æ–±—Ä–∞—Ç–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É –¥–ª—è $X$. –°—á–∏—Ç–∞–π—Ç–µ, —á—Ç–æ –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ $X$ –∏–º–µ–µ—Ç –ø–æ–ª–Ω—ã–π —Ä–∞–Ω–≥ (–º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–æ–∑–º–æ–∂–Ω—ã–π).

## 4. –ú–∞—Ç—Ä–∏—Ü–∞ –æ–±—ä–µ–∫—Ç-–ø—Ä–∏–∑–Ω–∞–∫.

**–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–π –≤–æ–ø—Ä–æ—Å 1.**

–ü—É—Å—Ç—å $X$ - –º–∞—Ç—Ä–∏—Ü–∞ –æ–±—ä–µ–∫—Ç-–ø—Ä–∏–∑–Ω–∞–∫ (—Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å $l \times F$ ), –¥–ª—è –∫–æ—Ç–æ—Ä–æ–π —Å–∏–Ω–≥—É–ª—è—Ä–Ω–æ–µ —Ä–∞–∑–ª–æ–∂–µ–Ω–∏–µ –∏–º–µ–µ—Ç –≤–∏–¥ $X=$ $V \sqrt{\Lambda} U^T$. –ü–æ—Å–ª–µ –ø–æ–Ω–∏–∂–µ–Ω–∏—è —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –¥–∞–Ω–Ω—ã—Ö —Å –ø–æ–º–æ—â—å—é –º–µ—Ç–æ–¥–∞ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç, –≤ –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω–æ–π –º–∞—Ç—Ä–∏—Ü–µ $\Lambda=\operatorname{diag}\left\{\lambda_1 \geq \cdots \geq \lambda_F\right\}$ –æ—Å—Ç–∞–≤–ª—è—é—Ç—Å—è —Ç–æ–ª—å–∫–æ $\tilde{F}$ –Ω–∞–∏–±–æ–ª—å—à–∏—Ö —Å–∏–Ω–≥—É–ª—è—Ä–Ω—ã—Ö —á–∏—Å–µ–ª: $\tilde{\Lambda}=\operatorname{diag}\left\{\lambda_1 \geq \cdots \geq \lambda_{\tilde{F}}\right\}$. –ü—Ä–∏ —ç—Ç–æ–º –¥–∞–Ω–Ω—ã–µ, –∫–∞–∫ –ø—Ä–∞–≤–∏–ª–æ, –º–æ–∂–Ω–æ –≤–æ—Å—Å—Ç–∞–Ω–æ–≤–∏—Ç—å —Ç–æ–ª—å–∫–æ —Å –Ω–µ–∫–æ—Ç–æ—Ä–æ–π –æ—à–∏–±–∫–æ–π: $\tilde{X}=V \sqrt{\tilde{\Lambda}} U^T \neq X$. –ü–æ–∫–∞–∂–∏—Ç–µ, —á—Ç–æ $L_2$ –Ω–æ—Ä–º–∞ –æ—à–∏–±–∫–∏ –≤—ã—Ä–∞–∂–∞–µ—Ç—Å—è —á–µ—Ä–µ–∑ —Å—É–º–º—É –ø–æ –æ—Å—Ç–∞–≤—à–∏–º—Å—è —Å–∏–Ω–≥—É–ª—è—Ä–Ω—ã–º —á–∏—Å–ª–∞–º:
$$
\frac{1}{l}\|X-\tilde{X}\|^2=\sum_{i=\tilde{F}+1}^F \lambda_i
$$

**–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–π –≤–æ–ø—Ä–æ—Å 2.**

–ü–æ–∫–∞–∂–∏—Ç–µ, —á—Ç–æ —Å–∏–Ω–≥—É–ª—è—Ä–Ω—ã–π –≤–µ–∫—Ç–æ—Ä –º–∞—Ç—Ä–∏—Ü—ã $X$, –æ—Ç–≤–µ—á–∞—é—â–∏–π –Ω–∞–∏–±–æ–ª—å—à–µ–º—É —Å–∏–Ω–≥—É–ª—è—Ä–Ω–æ–º—É —á–∏—Å–ª—É, —è–≤–ª—è–µ—Ç—Å—è —Ä–µ—à–µ–Ω–∏–µ–º –∑–∞–¥–∞—á–∏

$$
\boldsymbol{u}=\operatorname{argmax}_{|| \boldsymbol{u} \|=1}(X \boldsymbol{u})^2,
$$

–≥–¥–µ –ø–æ–¥—Ä–∞–∑—É–º–µ–≤–∞–µ—Ç—Å—è –º–∞—Ç—Ä–∏—á–Ω–æ–µ —É–º–Ω–æ–∂–µ–Ω–∏–µ $X$ –Ω–∞ $\boldsymbol{u}$.


**–ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –∑–∞–¥–∞–Ω–∏–µ.**
